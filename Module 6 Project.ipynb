```python
df.loc[df['TaxRate'] < 0, 'TaxRate'] *= -1
df.loc[df['ExtendedPrice'] < 0, 'ExtendedPrice'] *= -1
```

```python
df.drop(columns=['LastEditedWhen'],inplace = True)
```

```python
df[df['StockItemID'].isnull() & df['Description'].isnull()]
```

```python
df.dropna(subset=['StockItemID', 'Description'], how='all', inplace=True)
```

```python
df.shape
```

```python
df.describe()
```

```python
df['TaxAmount'] = df['UnitPrice'] * df['Quantity'] * df['TaxRate'] / 100
```

```python
df.describe()
```

```python
df['ExtendedPrice'] = df['UnitPrice'] * df['Quantity'] + df['TaxAmount']
```

```python
average_ep = df['ExtendedPrice'].mean()
std_dev_ep = df['ExtendedPrice'].std()
average_qty = df['Quantity'].mean()
std_dev_qty = df['Quantity'].std()
```

```python
df['Z-Score'] = (df['ExtendedPrice'] - average_ep) / std_dev_ep
df['Z-Scoreq'] = (df['Quantity'] - average_qty) / std_dev_qty
df.head(1)
```

```python
df.loc[df['Z-Scoreq'] > 3]
```

```python
df.loc[df['Z-Score'] > 3]
```

```python
df[df.duplicated(subset=['InvoiceID', 'InvoiceLineID'])]
```

```python
df.drop_duplicates(subset=['InvoiceID', 'InvoiceLineID'], keep='first', inplace=True)
```
