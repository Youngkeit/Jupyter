```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv(r"C:\Users\Barbara Youngkeit\Desktop\SQL\CSV\housing.csv")
```

```python
df.head(1)
```

```python
# Question 1) Create a scatter plot that shows land size on the x-axis and price on the y-axis
```

```python
sns.scatterplot(x='Landsize', y='Price', data=df, hue='Type')
plt.title("Property Price for Landsize")
plt.xlabel("Size of Land")
plt.ylabel("Price")

plt.show()
```

```python
# Question 3) Create a line chart that shows the year built along the x-axis and the average price per year on the y-axis
```

```python
df.sort_values(by="YearBuilt", inplace=True)
```

```python
avg_prices = df.groupby('YearBuilt').agg({'Price': 'mean'})
```

```python
sns.lineplot(x='YearBuilt', y='Price', data=avg_prices)
plt.title("Average Property Price per Year")
plt.xlabel("Year Built")
plt.ylabel("Average Price")
plt.show()
```

```python
# Question 5) Create a bar chart that shows the region on the x-axis and the average price of the property on the y-axis. 
```

```python
avg_price_prop = df.groupby('Regionname').agg({'Price': 'mean'})
```

```python
plt.figure(figsize=(18, 8))
sns.barplot(x = 'Regionname', y = 'Price', data = avg_price_prop)
plt.title("Average Property Price by Region")
plt.xlabel("Region")
plt.ylabel("Average Price")
plt.show()
```

```python
# Question 7) Create a histogram of property prices.
```

```python
sns.histplot(x='Price', data=df)
plt.show()
```

```python
# Question 9) Create a box plot that shows the years that each property was built on the x-axis and the region on the y-axis
```

```python
df.boxplot(column='YearBuilt', by='Regionname', vert=False)
plt.title('Years Property by Region')
plt.xlabel('Year Built')
plt.ylabel('Region')
plt.show()
```
